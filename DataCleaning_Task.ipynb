# Dirty cafe sales

```python
# importing necessary libraries
import pandas as pd
import numpy as np
```

```python
# loading the data
df = pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/dirty_cafe_sales.csv")
df.info()
```

```python
# replacing unwanted data "ERROR","UNKNOWN"
df.replace(["ERROR","UNKNOWN"],np.nan, inplace=True)
df['Item'].unique()
```

```python
# Handling null values in item column
df = df.dropna(subset=['Item'])
df["Item"].unique()
```

```python
df.info()
```

```python
# handling null values in Quantity column
freq=df['Quantity'].mode()[0]
df['Quantity']=df['Quantity'].fillna(freq)
df["Quantity"].unique()
```

```python
# handling null values in Price per unit column
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'])
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].mean())
df['Price Per Unit'] = df['Price Per Unit'].round(2)
df["Price Per Unit"].unique()
```

```python
# Total Spent col
df["Total Spent"].unique()
```

```python
# handling null values in Total Spent column
df['Total Spent'] = pd.to_numeric(df['Total Spent'])
df['Total Spent'] = df['Total Spent'].fillna(df['Total Spent'].mean())
df['Total Spent'] = df['Total Spent'].round(2)
df["Total Spent"].unique()
```

```python
# Payment Method col
df["Payment Method"].unique()
```

```python
# handling null values in Payment Method column
df['Payment Method'] = df['Payment Method'].replace({pd.NA:np.nan})
df['Payment Method'] = df['Payment Method'].fillna(df['Payment Method'].mode()[0])
df["Payment Method"].unique()
```

```python
# Location col
df["Location"].unique()
```

```python
# handling null values in Location column
df['Location'] = df['Location'].replace({pd.NA:np.nan})
df['Location'] = df['Location'].fillna(df['Location'].mode()[0])
df["Location"].unique()
```

```python
# Dropping null values
df.dropna(inplace=True)
df["Transaction Date"].unique()
```

```python
df.info()
```

```python
# dropping duplicate rows
df.drop_duplicates(inplace=True)
df.shape
```

```python
# resetting the index
df.reset_index(inplace=True)
# dropping unwanted columns
df = df.drop("index",axis=1)
df = df.drop("level_0",axis=1)
df
```
