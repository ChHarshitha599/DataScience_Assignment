```python
import numpy as np
```

# Part 1: Creating and Reshaping Arrays

```python
# 1. Create Arrays:
arr1 = np.linspace(0,50,10)
print(arr1)
arr2 = np.random.randint(5,20,9)
arr2=arr2.reshape(3,3)
print(arr2)
# 2. Array Dimensions:
print("dimension of arr1: ",arr1.ndim)
print("dimension of arr2: ",arr2.ndim)
print("shape of arr1: ",arr1.shape)
print("shape of arr2: ",arr2.shape)
# 3. Reshaping:
arr1=arr1.reshape(2,5)
print(arr1)
arr1=arr1.reshape(-1)
print(arr1)
```

# Part 2: Indexing and Slicing

```python
# 4. Indexing:
print(arr2[1,2])
print(arr2[:2,1:])
# 5. Slicing:
print(arr1[:4])
print(arr1[::-1])
```

#Part 3: Array Operations

```python
# 6. Array Arithmetic:
arr3 = np.arange(1,11)
print(arr3)
print(arr1+arr3)
print(arr1*arr3)
print(arr3-5)
```

# Part 4: Random Number Generation and Copying  

```python
# 7. Random Numbers:
arr4=np.random.rand(5)
print(arr4)
arr5=np.random.randint(-1,1,5)
print(arr5)
```
